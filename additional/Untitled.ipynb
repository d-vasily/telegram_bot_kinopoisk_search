```python
def get_free_proxies():
    url1 = "https://free-proxy-list.net/"
    # посылаем HTTP запрос и создаем soup объект
    soup = bs(requests.get(url1).content, "html.parser")
    proxies = []
    i = 0
    print('in get free proxies')
    proxylisttable = soup.find("table", {"class": "table table-striped table-bordered"}).find_all("tr")[1:]
    for row in proxylisttable:
        i += 1
        if i > 20:
            print('proxies ', len(proxies))
            break
        tds = row.find_all("td")
        try:
            ip = tds[0].text.strip()
            port = tds[1].text.strip()
            host = f"{ip}:{port}"
            proxies.append(host)
        except IndexError:
            continue
    return proxies
```
